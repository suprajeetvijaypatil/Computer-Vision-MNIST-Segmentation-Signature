```

# Handwritten Digit Classification using CNN (Google Colab Version)
# ---------------------------------------------------------------
# This notebook loads MNIST CSV files, prepares the dataset,
# builds a Convolutional Neural Network, trains it, and evaluates accuracy.

import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models

# ---------------------------------------------------------------
# 1. Load dataset
# ---------------------------------------------------------------
# NOTE: Upload mnist_train.csv and mnist_test.csv into Colab first
# using the left-side file browser or with the upload code below:
#
# from google.colab import files
# uploaded = files.upload()

train_df = pd.read_csv("mnist_train.csv")   # adjust path if needed
test_df = pd.read_csv("mnist_test.csv")     # adjust path if needed

# ---------------------------------------------------------------
# 2. Split features and labels
# ---------------------------------------------------------------
X_train = train_df.iloc[:, 1:].values
y_train = train_df.iloc[:, 0].values

X_test = test_df.iloc[:, 1:].values
y_test = test_df.iloc[:, 0].values

# ---------------------------------------------------------------
# 3. Reshape for CNN input (28x28 grayscale images)
# ---------------------------------------------------------------
X_train = X_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0

# ---------------------------------------------------------------
# 4. Build a simple CNN model
# ---------------------------------------------------------------
model = models.Sequential([
    # First convolution + pooling
    layers.Conv2D(32, (3, 3), activation="relu", input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    # Second convolution + pooling
    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),

    # Flatten and fully-connected layers
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")   # 10 digits (0–9)
])

# ---------------------------------------------------------------
# 5. Compile the model
# ---------------------------------------------------------------
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Show model summary
model.summary()

# ---------------------------------------------------------------
# 6. Train the model
# ---------------------------------------------------------------
history = model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.1
)

# ---------------------------------------------------------------
# 7. Evaluate on test set
# ---------------------------------------------------------------
loss, acc = model.evaluate(X_test, y_test, verbose=2)
print("Test Accuracy:", acc)
```
